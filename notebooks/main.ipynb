#  Step 3 — Document Chunking

##  Purpose

Large documents cannot be directly passed into LLMs because:
- context windows are limited,
- token cost increases,
- retrieval becomes inefficient.

Therefore documents are split into smaller semantic chunks.

#  Why Chunking Matters

Without chunking:
- retrieval quality decreases,
- irrelevant information increases,
- semantic precision weakens.

Chunking improves:
- contextual retrieval,
- semantic relevance,
- retrieval accuracy.


#  Mathematical Foundation

Suppose:
- document length = \(L\)
- chunk size = \(C\)
- overlap = \(O\)

Then approximate chunk count becomes:

\[
N \approx \frac{L}{C-O}
\]


#  Why Overlap is Important

Overlap preserves semantic continuity between chunks.

Without overlap:
- sentences may split incorrectly,
- context may be lost.

Overlap improves:
- coherence,
- retrieval quality,
- semantic preservation.


#  Recursive Splitting

The RecursiveCharacterTextSplitter recursively splits:
1. paragraphs,
2. sentences,
3. words,
4. characters.

This minimizes semantic fragmentation.


# Industry Standard

Typical RAG chunking:
- chunk size = 500–1500
- overlap = 10–30%

This balances:
- retrieval precision,
- token efficiency,
- semantic completeness.

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_documents(documents)

print("Total chunks:", len(chunks))
````

will appear near each other.


#  Similarity Computation

\[
\cos(\theta)=\frac{A \cdot B}{||A|| ||B||}
\]

measures semantic closeness.

#  Step 8 — Building Conversational Memory from Scratch

##  Purpose

This step implements a custom conversational memory system.

The chatbot stores:
- previous user messages,
- previous assistant responses,
- conversational history.

This enables:
- contextual continuity,
- multi-turn reasoning,
- follow-up understanding.

#  Why Memory Matters

Without memory:
- chatbot becomes stateless,
- follow-up questions fail,
- conversation loses coherence.

Example:

```text
User: What is GPT-4?
User: Who created it?
```

Without memory:
- second question becomes ambiguous.

With memory:
- chatbot understands:
```text
"it" = GPT-4
```


#  Mathematical Perspective

Conversation state:

\[
H = \{(u_1,a_1),(u_2,a_2),...,(u_n,a_n)\}
\]

Where:
- \(u_i\) = user message
- \(a_i\) = assistant response

Generation becomes:

\[
P(y|q,D,H)
\]

Where:
- \(q\) = current query
- \(D\) = retrieved documents
- \(H\) = conversation history


#  Why Build Memory from Scratch?

Modern LangChain versions frequently change memory APIs.

Custom memory provides:
- full control,
- version stability,
- debugging simplicity,
- framework independence.


#  Real-World Relevance

Conversational memory is used in:
- ChatGPT,
- customer support bots,
- AI copilots,
- personal assistants,
- enterprise chat systems.

````python
class ScratchMemory:
    def __init__(self, max_turns=5):
        self.max_turns = max_turns
        self.messages = []

    def save(self, user, assistant):
        self.messages.append({
            "role": "user",
            "content": user
        })

        self.messages.append({
            "role": "assistant",
            "content": assistant
        })

        self.messages = self.messages[-self.max_turns * 2:]

    def get_history(self):
        return "\n".join(
            f"{m['role']}: {m['content']}"
            for m in self.messages
        )

    def clear(self):
        self.messages = []
````

##  Purpose

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
{system_prompt}

Conversation History:
{history}

Context:
{context}

User Question:
{question}
""")

def rag_chat(question):

    # Retrieve docs
    docs = retriever.invoke(question)

    # Build context
    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    # Load memory
    history = memory.get_history()

    # Format prompt
    final_prompt = prompt.format(
        system_prompt=system_prompt,
        history=history,
        context=context,
        question=question
    )

    # Generate response
    response = llm.invoke(final_prompt)

    answer = response.content

    # Save memory
    memory.save(question, answer)

    return {
        "answer": answer,
        "source_documents": docs
    }
````

\cos(\theta)=\frac{A \cdot B}{||A|| ||B||}
\]

measures semantic closeness.

LLM reasoning uses self-attention:

\[
Attention(Q,K,V)=softmax\left(\frac{QK^T}{\sqrt{d_k}}\right)V
